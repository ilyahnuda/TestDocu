```python
import matplotlib.pyplot as plt
from matplotlib import image
import os
import math
```

```python
Abs_path = 'D:/PythonProjects/TestDocuSketch/plots/'
```

```python
with os.scandir(Abs_path) as files:
    path_all_plots = [file.path for file in files if file.name.endswith('.png')]
```

```python
path_all_plots
```

```python
rows = math.ceil(len(path_all_plots)/3)
figure, axis = plt.subplots(rows, 3,figsize=(22, 22))

axis = axis.flatten()
for i,plot in enumerate(path_all_plots):
    data = image.imread(plot)
    axis[i].imshow(data)
    axis[i].xaxis.set_visible(False)
    axis[i].yaxis.set_visible(False)
```

Первый график является матрицей ошибки, благодаря которой видна эффективность модель. В данном случае все данные были предсказаны верно. С 2 по 4 графики отображают отношение между показателями у пола и у потолков. На 2 и 3 видна что зависимость имеет линейный вид. Что касается последних 3, то они показывают какие распределение показателей mean, min, max для различных значений target метки.
