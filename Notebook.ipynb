```python
import matplotlib.pyplot as plt
from matplotlib import image
import os
import math
```

```python
Abs_path = 'D:/PythonProjects/TestDocuSketch/plots/'
```

```python
with os.scandir(Abs_path) as files:
    path_all_plots = [file.path for file in files if file.name.endswith('.png')]
```

```python
path_all_plots
```

```python
rows = math.ceil(len(path_all_plots)/3)
figure, axis = plt.subplots(rows, 3,figsize=(22, 22))

axis = axis.flatten()
for i,plot in enumerate(path_all_plots):
    data = image.imread(plot)
    axis[i].imshow(data)
    axis[i].xaxis.set_visible(False)
    axis[i].yaxis.set_visible(False)
```
